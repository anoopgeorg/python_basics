```python
def get_customer_reviews(page_product_link):
    try:
        open_prod_page = bs(urlopen(page_product_link),'html.parser')
    except Exception as e:
        print(e)
    cust_rev_boxes = open_prod_page.find_all("div",{"class":"a-section review aok-relative"})
    return page_product_link,cust_rev_boxes
```

```python
def get_customer_details(cust_review_boxes): 
    customer_review_list = []
    for customer in cust_review_boxes:
        customer_uname = ''
        customer_rating = ''
        customer_sentiment = ''
        customer_review_l = ''
        try:
            #print(customer.find("div",{"class":"a-profile-content"}).span.text)
            customer_uname = customer.find("div",{"class":"a-profile-content"}).span.text
            #print(customer.find("i",{"data-hook":"review-star-rating"}).span.text)
            customer_rating = customer.find("i",{"data-hook":"review-star-rating"}).span.text
            #print(customer.find("a",{"class":"a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"}).find_all("span"))
            # Below code to pick review summary from span tag as it does not have any identifier
            span_tags = customer.find("a",{"class":"a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"}).find_all("span")
            sentiment_span = [span for span in span_tags if not span.has_attr('class')]
            customer_sentiment = sentiment_span[0]
            #print(customer.find("div",{"class":"a-expander-content reviewText review-text-content a-expander-partial-collapse-content"}).span.text)
            customer_review_l = customer.find("div",{"class":"a-expander-content reviewText review-text-content a-expander-partial-collapse-content"}).span.text
            customer_review_details = {
                "customer_unam" : customer_uname,
                "customer_rating" : customer_rating,
                "customer_sentiment" : customer_sentiment,
                "customer_review_l" : customer_review_l            
            }

        except Exception as e:
            print(e)
        return customer_review_details
```

```python
get_customer_details(cust_rev_boxes)
```

```python
product_review_details = []
for product_link in page_product_links:
    product_review = {}
    product_link, customer_rev_boxes = get_customer_reviews(product_link)
    customer_review_details = get_customer_details(customer_rev_boxes)
    product_review = {
        "product_link" : product_link,
        "customer_reviews" : customer_review_details
    }
    product_review_details.append(product_review)
```
